```python
#data cleaning and preparation
#handle missing data
#identify columns with many missing values - columns with the highest portion of missing values
missing_percent = (df.isnull().sum()/len(df))*100
print(missing_percent.sort_values(ascending=False).head(10))
```

```python
# Drop columns with more than 85% missing values
threshold = 85
df_cleaned = df.loc[:, (df.isnull().sum() / len(df) * 100) < threshold]

# Drop rows where critical fields are missing
df_cleaned = df_cleaned.dropna(subset=["title", "abstract", "publish_time"])

# Fill missing authors and doi
df_cleaned["authors"] = df_cleaned["authors"].fillna("Unknown")
df_cleaned["doi"] = df_cleaned["doi"].fillna("No DOI")

# Normalize text
df_cleaned["title"] = df_cleaned["title"].str.lower()
df_cleaned["abstract"] = df_cleaned["abstract"].str.lower()

# Drop duplicates based on title + abstract
df_cleaned = df_cleaned.drop_duplicates(subset=["title", "abstract"])

# Reset index
df_cleaned = df_cleaned.reset_index(drop=True)

# Save cleaned version
df_cleaned.to_csv("metadata_cleaned.csv", index=False)
```

```python
# Convert publish_time to datetime
df_cleaned["publish_time"] = pd.to_datetime(df_cleaned["publish_time"], errors="coerce")

# Extract year from publish_time
df_cleaned["year"] = df_cleaned["publish_time"].dt.year

# Count words in abstract
df_cleaned["abstract_word_count"] = df_cleaned["abstract"].apply(
    lambda x: len(str(x).split())
)
#create title word count column
df_cleaned["title_word_count"] = df_cleaned["title"].apply(
    lambda x: len(str(x).split())
)
#first author extraction
df_cleaned["first_author"] = df_cleaned["authors"].apply(
    lambda x: str(x).split(",")[0] if x != "Unknown" else "Unknown"
)

```

```python
#most frequent words in titles(simple word frequency)
from collections import Counter
import re

# Combine all titles into one string
all_titles = " ".join(df_cleaned["title"].dropna())

# Basic cleaning: remove punctuation and split into words
words = re.findall(r"\b\w+\b", all_titles.lower())

# Count word frequencies
word_freq = Counter(words)

# Remove very common stopwords
stopwords = {"the", "of", "and", "in", "to", "for", "a", "on", "with", "an"}
filtered_freq = {word: count for word, count in word_freq.items() if word not in stopwords}

# Top 10 most frequent words
top_words = Counter(filtered_freq).most_common(10)
top_words
```

```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the cleaned dataset
df = pd.read_csv("metadata_cleaned.csv", low_memory=False)

# Convert publish_time to datetime if not already
df["publish_time"] = pd.to_datetime(df["publish_time"], errors="coerce")
df["year"] = df["publish_time"].dt.year

# --- Streamlit App ---
st.title("CORD-19 Dataset Explorer ")
st.markdown("Explore COVID-19 research metadata interactively.")

# Show a sample of data
st.subheader(" Sample Data")
st.dataframe(df.head())

# Sidebar filters
st.sidebar.header("Filters")
year_filter = st.sidebar.slider("Select Publication Year", 
                                int(df["year"].min()), 
                                int(df["year"].max()), 
                                (2019, 2021))

journal_filter = st.sidebar.selectbox("Select Journal", 
                                      options=["All"] + list(df["journal"].dropna().unique()))

# Apply filters
filtered_df = df[(df["year"] >= year_filter[0]) & (df["year"] <= year_filter[1])]
if journal_filter != "All":
    filtered_df = filtered_df[filtered_df["journal"] == journal_filter]

st.write(f"Showing {len(filtered_df)} papers after filtering.")

# --- Visualization 1: Publications Over Time ---
st.subheader(" Publications Over Time")
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(data=filtered_df, x="year", ax=ax, palette="Blues_r", order=sorted(filtered_df["year"].dropna().unique()))
plt.xticks(rotation=45)
st.pyplot(fig)

# --- Visualization 2: Top Journals ---
st.subheader(" Top Publishing Journals")
top_journals = filtered_df["journal"].value_counts().head(10)
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(y=top_journals.index, x=top_journals.values, palette="viridis", ax=ax)
st.pyplot(fig)

# --- Visualization 3: Word Cloud of Titles ---
st.subheader(" Word Cloud of Titles")
text = " ".join(title for title in filtered_df["title"].astype(str))
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
fig, ax = plt.subplots(figsize=(10,5))
ax.imshow(wordcloud, interpolation="bilinear")
ax.axis("off")
st.pyplot(fig)

# --- Visualization 4: Papers by Source ---
st.subheader(" Paper Sources Distribution")
top_sources = filtered_df["source_x"].value_counts().head(10)
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=top_sources.index, y=top_sources.values, palette="magma", ax=ax)
plt.xticks(rotation=45, ha="right")
st.pyplot(fig)
```
